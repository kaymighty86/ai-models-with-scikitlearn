# Cleaning Chronic Heart Disease Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/CKD - converted.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.drop("Unnamed: 25", axis = 1, inplace = True)
df.head()
```

```python
def get_from_df(receivable):
    if(receivable == "?"):
        return None
    else: 
        return receivable

df = df.map(get_from_df)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
df.isna().sum()
```
